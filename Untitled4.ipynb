```python
def Great_Wrapping():
    Dabao_data = {
        '定位': '在側胸部，腋中線上，當第6肋間隙處。',
        '解剖': '在第六肋間隙，前鋸肌中；有胸背動、靜脈及第六肋間動、靜脈；布有第六肋間神經，當胸長神經直系的末端。',
        '主治': '氣喘，胸脅病，全身疼痛，四肢無力。',
        '配伍': '配足三里治四肢無力。',
        '刺灸法': '斜刺或向後平刺0.5～0.8寸。'
    }
    return Dabao_data

def Supreme_White():
    Taibai_data = {
        '定位': '在足內側緣，當足大趾本節（第1跖骨關節）後下方赤白肉際凹陷處。',
        '解剖': '在拇展肌中；有足背靜脈網，足底內側動脈及足跗內側動脈分支；布有隱神經及腓淺神經分支。',
        '主治': '胃痛，腹脹，腸鳴，泄瀉，便秘，痔漏，腳氣，體重節痛。',
        '配伍': '配中脘、足三里治胃痛。',
        '刺灸法': '直刺0.5～0.8寸。'
    }
    return Taibai_data

def Grandfather_Grandson():
    Gongsun_data = {
        '定位': '在足內側緣，當第一跖骨基底部的前下方。',
        '解剖': '在拇展肌中；有跗內側動脈分支及足背靜脈網；布有隱神經及腓淺神經分支。',
        '主治': '胃痛，嘔吐，腹痛，泄瀉，痢疾。',
        '配伍': '配中脘、內關治胃酸過多、胃痛。',
        '刺灸法': '直刺0.6～1.2寸。'
    }
    return Gongsun_data

def Dripping_Valley():
    Lougu_data = {
        '定位': '在小腿內側，當內踝尖與陰陵泉的連線上，距內踝尖6寸，脛骨內側緣後方。',
        '解剖': '在脛骨後緣與比目魚肌之間，深層有屈趾長肌；有大隱靜脈，肢後動、靜脈；有小腿內側皮神經，深層內側後方有脛神經。',
        '主治': '腹脹，腸鳴，小便不利，遺精，下肢痿痹。',
        '配伍': ' 配足三里治腹脹腸鳴。',
        '刺灸法': '直刺1～1.5寸。'
    }
    return Lougu_data

def Spleen_Meridian_of_Foot_Taiyin():
    acupoints_info = {
        '大包穴': Great_Wrapping(),
        '太白穴': Supreme_White(),
        '公孫穴': Grandfather_Grandson(),
        '漏谷穴': Dripping_Valley(),
        # ... 其他穴位
    }
    return acupoints_info

def get_acupoint_info(acupoint_name, info_category=None):
    acupoints_info = Spleen_Meridian_of_Foot_Taiyin()

    acupoint_data = acupoints_info.get(acupoint_name, None)
    if acupoint_data is None:
        return "未找到指定的穴位信息。"

    if info_category:
        return acupoint_data.get(info_category, "未找到指定類別的信息。")

    return acupoint_data

def query_acupoint():
    while True:
        print("\n--- 查詢系統 ---")
        acupoint_name = input("請輸入您要查詢的穴位名稱（輸入'退出'結束查詢）:\n(大包穴、太白穴、公孫穴、漏谷穴選擇其一owo):")
        if acupoint_name == "退出":
            print("\n--- 已退出查詢 ---")
            print("感謝使用本系統，祝你好運owo")
            break

        acupoint_info = get_acupoint_info(acupoint_name)
        if "未找到" in acupoint_info:
            print(f"\n{acupoint_info}\n")
            continue

        while True:
            print(f"\n找到穴位：{acupoint_name}")
            print("可查詢的信息類別有：定位、解剖、主治、配伍、刺灸法")
            info_category = input("請輸入您要查詢的信息類別（輸入'全部'顯示所有信息，或輸入'返回'返回上一層）:")

            if info_category == "返回":
                print("\n返回上一層...\n")
                break
            elif info_category == "全部":
                print(f"\n{acupoint_name} - 全部信息:")
                for category, detail in acupoint_info.items():
                    print(f"{category}: {detail}")
                print()  # 為了排版整潔owo
            elif info_category in ["定位", "解剖", "主治", "配伍", "刺灸法"]:
                category_info = get_acupoint_info(acupoint_name, info_category)
                print(f"\n{acupoint_name} - {info_category}:\n{category_info}\n")
            else:
                print("\n輸入的信息類別不正確，請重新輸入。\n")

            # 再次查詢或返回
            while True:
                next_action = input("輸入'再查詢'以查詢此穴位的其他信息，或輸入'返回'返回穴位選擇：")
                if next_action in ["再查詢", "返回"]:
                    break
                else:
                    print("請輸入有效的指令（'再查詢' 或 '返回'）。")
            if next_action == "返回":
                print("\n返回穴位選擇...")
                break
# 啟動查詢
query_acupoint()
```
